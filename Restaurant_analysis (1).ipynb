```python
top_10 = df['Item Name'].value_counts().nlargest(10)
sns.barplot(x= top_10.values, y= top_10.index)
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top_10')
```

```python
top_cat = df['Category'].value_counts().nlargest(5)
sns.barplot(x=top_cat.values, y= top_cat.index)
plt.xlabel('Count')
plt.ylabel('Cat type')
```

```python
top_item = df.groupby('Item Name')['Qty.'].sum().nlargest(10)
sns.barplot(x= top_item.values , y= top_item.index)
plt.xlabel('-----')
plt.ylabel('-----')
plt.xticks(rotation = 45)
plt.title('Top sellers')
plt.show()
```

```python
top1 = df.groupby('Item Name')['Final Total'].sum().nlargest(10)
sns.barplot(x= top1.values, y= top1.index)
plt.xlabel('------')
plt.ylabel('------')
```

```python
top2 = df.groupby('Category')['Discount'].mean().nlargest(5)
sns.barplot(x= top2.values, y= top2.index)
```

```python
top3 = df.groupby('Category')['Qty.'].sum().nlargest(5)
sns.barplot(x= top3.values, y= top3.index)
```

```python
top4 = df.groupby('Category')['Final Total'].sum().nlargest(10)
sns.barplot(x= top4.values, y= top4.index)
```

```python
top5 = df.groupby(['Invoice No.','Item Name'])['Qty.'].sum().nlargest(10)
top5.plot(kind = 'bar',color = 'skyblue')
plt.show()
```
